```python
# Setup path
import sys
sys.path.insert(0, '..')

from backend.rag_service.core.rag_engine import RAGEngine, RAGConfig
from backend.rag_service.config.settings import settings

print("✓ Modules loaded")
print(f"✓ Settings loaded - API key present: {bool(settings.gemini.api_key)}")
```

```python
# Configure and test RAG query
question = "ما هي طريقة صلاة الخوف"

print("="*60)
print(f"Query: {question}")
print("="*60)

config = RAGConfig(
    top_k=3,
    use_linked_retrieval=False,  # Disable for initial test
    include_citations=True,
    citation_format="islamic_scholarly",
)

engine = RAGEngine(config=config, provider_name="gemini")
response = engine.query(question)

print("\nAnswer:")
print(response.answer)
print("\nCitations:")
for i, citation in enumerate(response.citations, 1):
    print(f"[{i}] {citation.get('formatted', citation.get('source', 'Unknown'))}")
```
